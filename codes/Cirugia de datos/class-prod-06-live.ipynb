## Dummy Variables --> Variables categóricas a numéricas

```python
import pandas as pd
import numpy as np
```

```python
MAINPATH = r'C:\Users\ThunderXForce\Desktop'
```

```python
brand = ["Gigabyte", "MSI", "Lenovo", "Dell", "Acer"]
graphic_card = ["Gtx 1660 Ti", "Rtx 2080 Ti", "Integrated"]
memory = ["4gb", "8gb", "16gb", "32gb"]
monitor_type = ["TN", "VA", "IPS"]
monitor_res = ["HD", "FHD", "4K"]
storage = ["128gb", "256gb" , "500gb", "1tb"]
storage_type = ["HDD", "SSD"]
```

```python
n_rows = 100000
brand_data = np.random.choice(brand, size=n_rows)
graphic_card_data = np.random.choice(graphic_card, size=n_rows)
memory_data = np.random.choice(memory, size=n_rows)
monitor_type_data = np.random.choice(monitor_type, size=n_rows)
monitor_res_data = np.random.choice(monitor_res, size=n_rows)
storage_data = np.random.choice(storage, size=n_rows)
storage_type_data = np.random.choice(storage_type, size=n_rows)
```

```python
data = pd.DataFrame({
    'Brand': brand_data,
    'Graphic Card': graphic_card_data,
    'Memory': memory_data,
    'Monitor Type': monitor_type_data,
    'Monitor Res': monitor_res_data,
    'Storage': storage_data,
    'Storage Type': storage_type_data
})
```

```python
data.head()
```

```python
data.shape
```

```python
dummy_storage_type = pd.get_dummies(data['Storage Type'], prefix="disk")
dummy_storage_type
```

```python
data2 = data.drop(['Storage Type'], axis="columns")
data2
```

```python
new_computers = pd.DataFrame({
    'Brand': ['Asus', 'Lenovo', 'Samsung', 'Microsoft', 'Apple'],
    'Graphic Card': ['Gtx 1650 Ti', 'Gtx 1440 Ti', 'Integrated', 'Rtx 2060 Ti', 'Integrated'],
    'Memory': ['32gb', '32gb', '64gb', '128gb', '2tb'],
    'Monitor Type': ['TN', 'TN', 'VA', 'IPS', 'VA'],
    'Monitor Res': ['2K', '4K', 'FHD', 'HD', 'HD'],
    'Storage': ['32gb', '160gb', '512gb', '1tb', '200tb'],
    'Storage Type': ['HDD', 'SSD', 'SSD', 'HDD', 'SSD']
})
new_computers
```

```python
data3 = pd.concat([data, new_computers], axis="rows")
data3 = data3.reset_index().drop(['index'], axis="columns")
data3
```

## Método 2 --> join

```python
data.head()
```

```python
brands = data['Brand'].unique().tolist()
brands
```

```python
brands_df = pd.DataFrame({
    'Brand': brands, 
    'Brand Foundation Year': ['1975', '1984', '1999', '2010', '1980'],
    'Brand Capital': ['$10000000', '$5000000', '$100000000', '$1', '$9999999999999999999']
})
brands_df
```

```python
left = data.set_index('Brand')
right = brands_df.set_index('Brand')
combined_df = left.join(right, lsuffix="_comp", rsuffix="_brand")
combined_df
```
