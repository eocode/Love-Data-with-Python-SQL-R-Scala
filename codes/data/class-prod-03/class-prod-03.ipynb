# Data Wrangling

El Data Wrangling es la operación de transformar datos para hacerlos más apropiados para un trabajo posterior. La persona que realiza esta labor se denomina **data wrangler**.

El Data Wrangling incluye:
* Extracción de datos
* División de datos
* Visualización de datos
* Agregación de datos
* Depósito de datos (silos)

```python
## Pasajeros que tienen más de 30 años y sobrevivieron
new_df = data[(data["age"]>30) & (data["survived"]==1)]
new_df
```
