## Lectura manual de archivos

```python
data = open('./datasets/titanic.csv')
```

```python
columns = data.readline().strip().split(',')
columns
```

```python
n_columns = len(columns)
n_columns
```

```python
dataframe_dict = {column_name: ['Hola'] for column_name in columns}
dataframe_dict
```

```python
dataframe_dict['Sex']
```

```python
counter = 1
for line in data:
    values = line.strip().split(',')
    for i in range(len(columns)):
        dataframe_dict[columns[i]].append(values[i])
    counter += 1
    
print("El dataset tiene ",counter," filas y ",n_columns," columnas")
```
