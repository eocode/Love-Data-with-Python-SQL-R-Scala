## Dummy Variables --> Variables categóricas a numéricas

```python
import pandas as pd
import numpy as np
```

```python
MAINPATH = r"C:\Users\ThunderXForce\Desktop"
```

```python
brand = ["Gigabyte", "MSI", "Lenovo", "Dell", "Acer"] 
graphic_card = ["Gtx 1660 Ti", "Rtx 2080 Ti", "Integrated"]
memory = ["4gb", "8gb", "16gb"]
monitor_type = ["TN", "VA", "IPS"]
monitor_res = ["HD", "FHD", "4K"]
```

```python
n_rows = 100
brand_data = np.random.choice(brand, size=n_rows)
graphic_card_data = np.random.choice(graphic_card, size=n_rows)
memory_data = np.random.choice(memory, size=n_rows)
monitor_type_data = np.random.choice(monitor_type, size=n_rows)
monitor_res_data = np.random.choice(monitor_res, size=n_rows)
```

```python
data = pd.DataFrame(
    {
        'Brand': brand_data,
        'Graphic Card': graphic_card_data,
        'Memory': memory_data,
        'Monitor Type': monitor_type_data,
        'Monitor Resolution': monitor_res_data,
    }
)
```

```python
data.head()
```

```python
dummy_screen = pd.get_dummies(data['Monitor Resolution'], prefix="Screen")
dummy_screen
```

```python
data2 = data.drop(['Monitor Resolution'], axis=1)
data2
```

## Combinación de DataFrames

### Método 1 --> pd.concat

```python
data2 = pd.concat([data2, dummy_screen], axis=1)
data2
```

```python
new_computers = pd.DataFrame({
    'Brand': ['Lenovo', 'Intel', 'AMD', 'Lenovo', 'AMD'],
    'Graphic Card': ['Integrated', 'Rtx 2060 Ti', 'Gtx 1650 Ti', 'Integrated', 'Integrated'],
    'Memory': ['64gb', '128gb', '256gb', '512gb', '1tb'],
    'Monitor Type': ['VA', 'IPS', 'VA', 'TN', 'IPS'],
    'Monitor Resolution': ['4K', 'FHD', 'HD', 'HD', '4K']
})
new_computers
```

```python
data
```

```python
data3 = pd.concat([new_computers, data], axis=0)
data3 = data3.reset_index().drop(['index'], axis=1)
data3
```

### Método 2 --> join

```python
brands = data['Brand'].unique().tolist()
brands
```

```python
brands_df = pd.DataFrame({
    'Brand': brands,
    'Brand Foundation Year': ['1975', '1984', '1999', '2010', '1980'],
    'Brand Capital': ['$10000000', '$20000000', '$5000000', '$100000000', '$100']
})
brands_df
```

```python
left = data.set_index('Brand')
right = brands_df.set_index('Brand')
combined_df = left.join(right, lsuffix='_COMP', rsuffix='_BRAND').reset_index()
combined_df
```
