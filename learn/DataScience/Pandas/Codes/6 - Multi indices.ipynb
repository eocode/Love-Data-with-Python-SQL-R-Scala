```python
import pandas as pd
import numpy as np
```

```python
dir = './db/poblacion.csv'
dir
```

```python
df = pd.read_csv(dir)
df.head()
```

```python
df = df.drop(columns=['Country Code','Indicator Name','Indicator Code','2019'])
```

```python
df
```

```python
df = pd.melt(df, id_vars=["Country Name"], 
                     var_name="year", value_name="pob")
```

```python
df
```

```python
pd.options.display.float_format = '{:,.1f}'.format
```

```python
df['year'] = pd.Categorical(df['year'].apply(str))
```

```python
df.dtypes
```

```python
idx = df['Country Name'].isin(['Aruba', 'Colombia'])
idx
```

```python
df_sample = df[idx]
df_sample
```

```python
df_sample = df_sample.set_index(['Country Name','year']).sort_index()
df_sample
```
