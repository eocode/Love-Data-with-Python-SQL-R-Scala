```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('./db/GlobalLandTemperaturesByCountry.csv')
df
```

```python
df.shape
```

```python
df.info()
```

```python
df['Country'].unique()
```

```python
df['year'] = pd.to_datetime(df['dt'])
df['year']
```

```python
idx = df['year'] > pd.to_datetime('1970-01-01')
idx
```

```python
df = df[idx]
df
```

```python
df_t_avg = df.groupby([
    'Country',
    pd.Grouper(key='year', freq='1Y')
]).agg({'AverageTemperature':[np.mean, np.median]})
df_t_avg
```

```python
df_t_avg.xs('Mexico')
```

```python
df_t_avg.xs('Mexico')['c'].plot()
```

```python
d_t_md = df_t_avg['AverageTemperature'][['median']].reset_index()
d_t_md
```

```python
d_t_md['date'] = d_t_md['year'].dt.year
d_t_md['date']
```

```python
d_t_md.rename(columns={'median':'temperature'}, inplace=True)
d_t_md
```

```python
d_t_pivot = d_t_md.pivot_table(values='temperature',index='date', columns='Country')
d_t_pivot
```

```python
d_t_pivot.boxplot()
```

```python
df_agri = pd.read_csv('./db/API_AG.LND.AGRI.K2_DS2_en_csv_v2_1000207.csv',header=2)
df_fore = pd.read_csv('./db/API_AG.LND.FRST.K2_DS2_en_csv_v2_989381.csv',header=2)
df_elec = pd.read_csv('./db/API_EG.USE.ELEC.KH.PC_DS2_en_csv_v2_989401.csv',header=2)
df_co2e = pd.read_csv('./db/API_EN.ATM.CO2E.KT_DS2_en_csv_v2_989664.csv',header=2)
df_popu = pd.read_csv('./db/API_SP.POP.TOTL_DS2_en_csv_v2_988606.csv',header=2)
```

```python
df_popu
```

```python
range(1971,2015)
map(str, range(1971,2015))
cols= ['Country Name', 'Country Code'] + list(map(str, range(1971,2015)))
cols[:5]
```

```python
df_agri.loc[:,cols]
```

```python
df_agri.loc[:,cols].melt(id_vars=['Country Name','Country Code'])
```

```python
df_agri.loc[:,cols].melt(id_vars=['Country Name','Country Code']).rename(
    columns={'variable':'date',
             'Country Name':'Country',
             'Country Code':'name',
             'value':'agriculture'})
```

```python
def fun_format(df, col = 'agriculture'):
    return df.loc[:,cols].melt(id_vars=['Country Name','Country Code']).rename(
        columns={'variable':'date',
                 'Country Name':'Country',
                 'Country Code':'name',
                 'value': col})
```

```python
df_agri = fun_format(df_agri, col = 'agriculture')
df_fore = fun_format(df_fore, col = 'forest')
df_elec = fun_format(df_elec, col = 'electricprod')
df_co2e = fun_format(df_co2e, col = 'co2')
df_popu = fun_format(df_popu, col = 'population')
df_popu
```

```python
df_popu.info()
```

```python
df_popu['date'] = df_popu['date'].astype(float)
df_fore['date'] = df_fore['date'].astype(float)
df_agri['date'] = df_agri['date'].astype(float)
df_elec['date'] = df_elec['date'].astype(float)
df_co2e['date'] = df_co2e['date'].astype(float)
df_popu.info()
```

```python
df_merge = pd.merge(d_t_md[['Country','temperature','date']],
                   df_popu, on = ['Country', 'date'], how = 'inner'
                   )
df_merge
```

```python
df_merge = pd.merge(df_merge,
                    df_co2e,
                    on = ['Country','name','date'],
                    how = 'inner')

df_merge = pd.merge(df_merge,
                    df_elec,
                    on = ['Country','name','date'],
                    how = 'inner')

df_merge = pd.merge(df_merge,
                    df_agri,
                    on = ['Country','name','date'],
                    how = 'inner')

df_merge = pd.merge(df_merge,
                    df_fore,
                    on = ['Country','name','date'],
                    how = 'inner')
df_merge
```

```python
df_merge.dropna()
```

```python
df_climate = df_merge.dropna()
```
